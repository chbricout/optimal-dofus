Demonstration de quelques fonctions d'API avec dofusDB

# Graphe de précédence des quêtes de Dofus

```python
%load_ext autoreload
%autoreload 2
from dofusdb import model, api_loader, graph_creator, sql_loader,data_agg
```

## Suite de quêtes d'Incarnam

### Juste les quêtes de la catégorie Incarnam

```python
quests_inc = api_loader.load_quest_from_category(19)
data_agg.complete_quest_dict(quests_inc)
graph_creator.graph_from_quests("Incarnam", quests_inc, group_criterion=False, render_as=None)
```

### Inclus les quêtes de la trame principale

```python
quests_inc= api_loader.load_quest_from_category(19)
for idx,q in api_loader.load_quest_from_achievement(1378).items():
    quests_inc[idx]=q
data_agg.complete_quest_dict(quests_inc)
graph_creator.graph_from_quests("Incarnam-Principale", quests_inc, render_as=None)
```

## Suite de quêtes d'Astrub

```python
quests_astr= api_loader.load_quest_from_category(3)
data_agg.complete_quest_dict(quests_astr)
graph_creator.graph_from_quests("Astrub", quests_astr, render_as=None)
```

On voit au-dessus que toute une ligne de quête dépend en réalité de la classe du joueur, factoriser ces quêtes permettrait une présentation plus propre.
Pour cela on commence par les détecter :

```python
data_agg.detects_quests_with_class_cluster(quests_astr)
```

```python
data_agg.replace_class_dependent_quests(quests_astr)
graph_creator.graph_from_quests("Astrub without class criterion", quests_astr, render_as=None)
```
